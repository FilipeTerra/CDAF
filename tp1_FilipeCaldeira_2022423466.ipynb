```python
# Seleciona jogos do 1º Turno de acordo com a data
ds_alemao2122['Date'] = pd.to_datetime(ds_alemao2122['Date'], dayfirst=True)
ds_primeira_metade = ds_alemao2122.sort_values(by='Date').iloc[:len(ds_alemao2122) // 2]
tabela_1turno = {}

# Iterar sobre cada partida para calcular os dados
for _, row in ds_primeira_metade.iterrows():
    home_team, away_team = row['HT'], row['AT']
    home_goals, away_goals = row['HS'], row['AS']

    # Criar entrada para os times caso ainda não existam
    if home_team not in tabela_1turno:
        tabela_1turno[home_team] = {'Pontos': 0, 'Jogos': 0, 'Vitórias': 0, 'Empates': 0, 'Derrotas': 0, 'GP': 0, 'GC': 0, 'SG': 0}
    if away_team not in tabela_1turno:
        tabela_1turno[away_team] = {'Pontos': 0, 'Jogos': 0, 'Vitórias': 0, 'Empates': 0, 'Derrotas': 0, 'GP': 0, 'GC': 0, 'SG': 0}

    # Atualizar estatísticas dos times mandante e visitante
    tabela_1turno[home_team]['Jogos'] += 1
    tabela_1turno[away_team]['Jogos'] += 1

    tabela_1turno[home_team]['GP'] += home_goals
    tabela_1turno[home_team]['GC'] += away_goals
    tabela_1turno[home_team]['SG'] += home_goals - away_goals

    tabela_1turno[away_team]['GP'] += away_goals
    tabela_1turno[away_team]['GC'] += home_goals
    tabela_1turno[away_team]['SG'] += away_goals - home_goals

    # Definir resultado do jogo
    if home_goals > away_goals:  # Vitória do mandante
        tabela_1turno[home_team]['Vitórias'] += 1
        tabela_1turno[home_team]['Pontos'] += 3
        tabela_1turno[away_team]['Derrotas'] += 1
    elif home_goals < away_goals:  # Vitória do visitante
        tabela_1turno[away_team]['Vitórias'] += 1
        tabela_1turno[away_team]['Pontos'] += 3
        tabela_1turno[home_team]['Derrotas'] += 1
    else:  # Empate
        tabela_1turno[home_team]['Empates'] += 1
        tabela_1turno[away_team]['Empates'] += 1
        tabela_1turno[home_team]['Pontos'] += 1
        tabela_1turno[away_team]['Pontos'] += 1

# Criar DataFrame a partir do dicionário
df_1turno = pd.DataFrame.from_dict(tabela_1turno, orient='index')

# Adicionar coluna de nome dos times
df_1turno.insert(0, 'Clube', df_1turno.index)

# Ordenar pelo critério de desempate: Pontos, Vitórias, Saldo de Gols, Gols Pró
df_1turno = df_1turno.sort_values(by=['Pontos', 'Vitórias', 'SG', 'GP'], ascending=[False, False, False, False]).reset_index(drop=True)

# Exibir a tabela final
df_1turno.index = df_1turno.index + 1
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
print('Tabela 1º Turno Bundesliga 21/22\n')
print(df_1turno)
```

## Questão 4
- Utilizando os jogos da liga escolhida, use regressão de Poisson para criar um modelo de previsão de resultados, como visto nos slides em sala e no Soccermatics.
-- https://soccermatics.readthedocs.io/en/latest/gallery/lesson5/plot_SimulateMatches.html
- Dê print no sumário do ajuste
- Simule a partida entre o 1º e o 4º colocado, onde o 1º joga em casa. Primeiro, apresente a quantidade esperada de gols de cada time. Em seguida, apresente um grid com as probabilidades associadas a cada placar possível.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import seaborn
from scipy.stats import poisson,skellam
```

```python
# Criar DataFrames para os gols dos mandantes e visitantes
home_goals = ds_alemao2122[['HT', 'AT', 'HS']].assign(home=1).rename(
    columns={'HT': 'team', 'AT': 'opponent', 'HS': 'goals'})

away_goals = ds_alemao2122[['AT', 'HT', 'AS']].assign(home=0).rename(
    columns={'AT': 'team', 'HT': 'opponent', 'AS': 'goals'})

# Concatenar os dois conjuntos de dados
goal_model_data = pd.concat([home_goals, away_goals], ignore_index=True)

# Ajustar o modelo de regressão de Poisson
poisson_model = smf.glm(
    formula="goals ~ home + team + opponent",
    data=goal_model_data,
    family=sm.families.Poisson()
).fit()

# Exibir o resumo do modelo
print(poisson_model.summary())
```

Simule a partida entre o 1º e o 4º colocado, onde o 1º joga em casa. Primeiro, apresente a quantidade esperada de gols de cada time. Em seguida, apresente um grid com as probabilidades associadas a cada placar possível.

```python
# Selecionar os times
team_home = df_classificacao.iloc[0]['Clube']  # 1º colocado
team_away = df_classificacao.iloc[3]['Clube']  # 4º colocado

# Criar DataFrames para os gols dos mandantes e visitantes
home_goals = ds_alemao2122[['HT', 'AT', 'HS']].assign(home=1).rename(
    columns={'HT': 'team', 'AT': 'opponent', 'HS': 'goals'})

away_goals = ds_alemao2122[['AT', 'HT', 'AS']].assign(home=0).rename(
    columns={'AT': 'team', 'HT': 'opponent', 'AS': 'goals'})

# Função para prever gols esperados
def predict_goals(model, team, opponent, home=1):
    """Retorna a estimativa de gols esperados para um time em uma partida."""
    return np.exp(model.predict(pd.DataFrame({'team': [team], 'opponent': [opponent], 'home': [home]}))).item()

# Função para simular a matriz de probabilidades de placares
def simulate_match(model, home_team, away_team, max_goals=5):
    """Retorna a matriz de probabilidades de placares possíveis em um jogo."""
    lambda_home = predict_goals(model, home_team, away_team, home=1)
    lambda_away = predict_goals(model, away_team, home_team, home=0)

    prob_home = [poisson.pmf(i, lambda_home) for i in range(max_goals + 1)]
    prob_away = [poisson.pmf(i, lambda_away) for i in range(max_goals + 1)]

    return np.outer(prob_home, prob_away)

# Simular a partida
max_goals = 5
score_matrix = simulate_match(poisson_model, team_home, team_away, max_goals)

# Criar heatmap das probabilidades de placares
plt.figure(figsize=(8, 6))
sns.heatmap(score_matrix, annot=True, fmt=".3f", cmap="Reds",
            xticklabels=range(max_goals + 1), yticklabels=range(max_goals + 1))

# Ajustar eixo Y para começar do zero
plt.gca().invert_yaxis()

plt.xlabel(f'Gols de {team_away}')
plt.ylabel(f'Gols de {team_home}')
plt.title(f'Probabilidades de Placar - {team_home} x {team_away}')
plt.show()
```

```python
# Criar a tabela simulada baseada nos resultados esperados
def simulate_match(home, away):
    home_goals = poisson_model.predict(pd.DataFrame({'team': [home], 'opponent': [away], 'home': [1]})).values[0]
    away_goals = poisson_model.predict(pd.DataFrame({'team': [away], 'opponent': [home], 'home': [0]})).values[0]
    return round(home_goals), round(away_goals)

teams = list(set(ds_alemao2122['HT']).union(set(ds_alemao2122['AT'])))
simulated_table = pd.DataFrame(index=teams, columns=['Pts', 'J', 'V', 'E', 'D', 'GP', 'GC', 'SG'])
simulated_table[:] = 0

# Simular os jogos e preencher a tabela
for index, row in ds_alemao2122.iterrows():
    home, away = row['HT'], row['AT']
    home_goals, away_goals = simulate_match(home, away)

    # Atualizar estatísticas do time mandante
    simulated_table.at[home, 'J'] += 1
    simulated_table.at[home, 'GP'] += home_goals
    simulated_table.at[home, 'GC'] += away_goals
    simulated_table.at[home, 'SG'] = simulated_table.at[home, 'GP'] - simulated_table.at[home, 'GC']

    # Atualizar estatísticas do time visitante
    simulated_table.at[away, 'J'] += 1
    simulated_table.at[away, 'GP'] += away_goals
    simulated_table.at[away, 'GC'] += home_goals
    simulated_table.at[away, 'SG'] = simulated_table.at[away, 'GP'] - simulated_table.at[away, 'GC']

    # Definir vitórias, empates e derrotas
    if (home_goals > away_goals) and (home_goals - away_goals > 0.5):
        simulated_table.at[home, 'V'] += 1
        simulated_table.at[home, 'Pts'] += 3
        simulated_table.at[away, 'D'] += 1
    elif (away_goals > home_goals) and (away_goals - home_goals > 0.5):
        simulated_table.at[away, 'V'] += 1
        simulated_table.at[away, 'Pts'] += 3
        simulated_table.at[home, 'D'] += 1
    else:
        simulated_table.at[home, 'E'] += 1
        simulated_table.at[away, 'E'] += 1
        simulated_table.at[home, 'Pts'] += 1
        simulated_table.at[away, 'Pts'] += 1

# Ordenar a tabela simulada
simulated_table = simulated_table.sort_values(by=['Pts', 'V', 'SG', 'GP'], ascending=[False, False, False, False])

# Exibir a tabela simulada
print("\nTabela Simulada:")
print(simulated_table)
```
